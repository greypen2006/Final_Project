### Melt and Concatenate DataFrames

Transform `df1` and `df2` from year-month columns to rows, then combine them into a single DataFrame.


```python
# Melt the 2 dfs using the ids MajorText, MinorText and BoroughName, create column for the date and crime count
id_vars = ['MajorText', 'MinorText', 'BoroughName']
value_vars_df1 = [col for col in df1.columns if col not in id_vars]
df1_melted = pd.melt(df1, id_vars=id_vars, value_vars=value_vars_df1, var_name='YearMonth', value_name='CrimeCount')
df1_melted.head()
```

```python
# Same for df2
id_vars = ['MajorText', 'MinorText', 'BoroughName']
value_vars_df2 = [col for col in df2.columns if col not in id_vars]
df2_melted = pd.melt(df2, id_vars=id_vars, value_vars=value_vars_df2, var_name='YearMonth', value_name='CrimeCount')
df2_melted.head()
```

```python
# Concatenate the 2
combined_df = pd.concat([df1_melted, df2_melted], ignore_index=True)
combined_df.head()
```

```python
# Verifying data types and structure
combined_df.info()
```

### Extract and Filter by Year

Extract the year from the 'YearMonth' column and filter the combined DataFrame to include only data for the years 2017 to 2024. Taking first 4 characters and convert to integer type.


```python
combined_df['Year'] = combined_df['YearMonth'].astype(str).str[:4].astype(int)
combined_df.head()
```

```python
# Filter for 2017-2024
combined_df_filtered = combined_df[(combined_df['Year'] >= 2017) & (combined_df['Year'] <= 2024)]
combined_df_filtered.head()
```

### Filter by Boroughs

Filter the DataFrame to include only the specified boroughs: Camden, Islington, Hackney, Tower Hamlets, Greenwich, Southwark, Lambeth, Hammersmith and Fulham, Kensington and Chelsea, Westminster, and Brent.


```python
borough_list = ['Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich', 'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea', 'Westminster', 'Brent']
print("Borough list created.")
```

```python
combined_df_borough_filtered = combined_df_filtered[combined_df_filtered['BoroughName'].isin(borough_list)]
combined_df_borough_filtered.head()
```

### Convert Crime count to crime rate with population data

```python
#Continuous issues with tokenising data due to structure --> lead to importing data from google.colab directly (available as Population.csv in Raw Datasets folder)
```

```python
from google.colab import files
uploaded2 = files.upload()
```

```python
# Skip to Total population data
pop_df=pd.read_csv('Population.csv', skiprows=63, encoding='latin1')
pop_df = pop_df.rename(columns={'local authority: county / unitary (as of April 2023)': 'BoroughName'}) # rename for consistency
id_vars = ['BoroughName', 'mnemonic']
value_vars = [str(year) for year in range(2017, 2025)] # Select years to filter
pop_df_melted = pd.melt(pop_df, id_vars=id_vars, value_vars=value_vars, var_name='Year', value_name='Population') # melt together
pop_df_melted['Year'] = pop_df_melted['Year'].astype(int)
pop_df_melted
```

```python
# Merge on BoroughName, Year
combined_df_merged = pd.merge(combined_df_borough_filtered, pop_df_melted, on=['BoroughName', 'Year'], how='left')
# Calculate crime rate by dividing by population and *100 (percentage)
combined_df_merged['CrimeRate'] = (combined_df_merged['CrimeCount'] / combined_df_merged['Population']) * 100
combined_df_merged.head()
```

```python
# Save to upload to cleaned datasets
combined_df_merged.to_csv('crime_cleaned_final.csv', index=False)
```

```python
!mkdir -p data
!curl -L 'https://drive.google.com/uc?export=download&id=107i0YICDfbz9DSa66mgdiyFKtJPu1BpM&confirm=xxx' -o ./data/femalepopuncleaned.csv
```

```python
df_fpop=pd.read_csv('./data/femalepopuncleaned.csv')
df_fpop
```

```python
#remove column 'code'
df_fpop = df_fpop.drop(columns=['Code'])
#only look at London boroughs
df_fpop = df_fpop[df_fpop['Geography'] == 'London Borough']
df_fpop.head()
```

```python
#Keep relevant boroughs
boroughs_to_keep = ['Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich', 'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea', 'Westminster', 'Brent']
df_fpop = df_fpop[df_fpop['Name'].isin(boroughs_to_keep)]
df_fpop.head()
```

```python
#Keep relevant ages and boroughs
df_fpop.columns = df_fpop.columns.str.strip()
cols_to_keep = ['Name', 'All ages'] + [str(age) for age in range(18, 26)]
df_fpop = df_fpop[cols_to_keep]
df_fpop = df_fpop.rename(columns={'Name': 'Borough'})
df_fpop
```

```python
#Remove commas from dataset
age_cols = ['All ages'] + [str(age) for age in range(18, 26)]
df_fpop[age_cols] = (df_fpop[age_cols].replace(',', '', regex=True).apply(pd.to_numeric))
df_fpop
```

```python
df_fpop.to_csv('femalepop_by_borough_and_age_cleaned.csv', index=False)
```

```python
df_inc=pd.read_csv('./data/incomeuncleaned.csv', encoding='latin1')
df_inc
```

```python
#Drop rows with NaN
rows_to_drop = list(range(35, 51)) + [0, 1]
df_inc = df_inc.drop(index=rows_to_drop)
df_inc.head()
```

```python
#Remove the 'Code' column and the confidence intervals for each year (the 'Unnamed' columns)
df_inc = df_inc.drop(columns=['Code'])
df_inc = df_inc.drop(columns=[col for col in df_inc.columns if col.startswith('Unnamed')])
df_inc.head()
```

```python
#Rename borough column and keep only the 11 relevant boroughs
df_inc = df_inc.rename(columns={'Area': 'Borough'})
boroughs_to_keep = ['Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich', 'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea', 'Westminster', 'Brent']
df_inc = df_inc[df_inc['Borough'].isin(boroughs_to_keep)]
#Switch years with boroughs
df_inc = df_inc.T
df_inc.head()
```

```python
#Remove the index and make the table neater by adding a 'Year' column
df_inc = df_inc.reset_index()
df_inc.columns = df_inc.iloc[0]
df_inc = df_inc.drop(df_inc.index[0])
df_inc = df_inc.rename(columns={df_inc.columns[0]: 'Year'})
df_inc = df_inc.set_index('Year')
df_inc
```

```python
#Only keep years 2017 to 2024
df_inc.index = df_inc.index.astype(int)
cleaned_df_inc = df_inc.loc[2017:2024]
cleaned_df_inc
```
