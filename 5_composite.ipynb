```python
import pandas as pd
file1_id = "1Dy-eciO2EXLkgbmP1FRoQrYt9p_qkMj6"
file1_url = f"https://drive.google.com/uc?id={file1_id}"
df1 = pd.read_csv(file1_url)
df1.head()

file2_id = "10lDwtL9VawYDqiaM4hhP9ASP4PHSDf5H"
file2_url = f"https://drive.google.com/uc?id={file2_id}"
df2 = pd.read_csv(file2_url)
df2.head()
```

```python
# Create a new dataframe based on df1 for entries where MajorText is SEXUAL OFFENSES
df_sexual_offences = df1[df1['MajorText'] == 'SEXUAL OFFENCES']

# Merge the data for each year of each borough to get the total CrimeCount of SEXUAL OFFENCES for each borough
df_sexual_offences_summary = df_sexual_offences.groupby(['BoroughName'])['CrimeCount'].sum().reset_index()

df_sexual_offences_summary
```

```python
wp, wc = 0.5, 0.5

crime_score_calc = (
    df_sexual_offences_summary["CrimeCount"].max() - df_sexual_offences_summary["CrimeCount"]
) / (
    df_sexual_offences_summary["CrimeCount"].max() - df_sexual_offences_summary["CrimeCount"].min()
    )

df_sexual_offences_summary["crime_score"] = crime_score_calc

df_sexual_offences_summary
```

```python
df2["rent_score"] = (
    df2["Mean_Monthly_Rent"].max() - df2["Mean_Monthly_Rent"]
) / (
    df2["Mean_Monthly_Rent"].max() - df2["Mean_Monthly_Rent"].min()
)

df2["Composite_Score"] = wp * df2["rent_score"] + wc * df_sexual_offences_summary["crime_score"]

df2 = df2.sort_values("Composite_Score", ascending = False)

df2
```
