```python
from matplotlib import pyplot as plt
from matplotlib import pylab
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as mcolors
import matplotlib as mpl
from matplotlib import cm
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('income weekly by borough uncleaned.csv',encoding='cp1252')
df
```

```python
df = df.drop(df.index[35:51])
df = df.drop(df.index[0:1])
df.head()
```

```python
df = df.drop(columns=['Code'])
df = df.drop(columns=[col for col in df.columns if col.startswith('Unnamed')])
df.head()
```

```python
df = df.rename(columns={'Area': 'Borough'})
boroughs_to_keep = ['Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich', 'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea', 'Westminster', 'Brent']
df = df[df['Borough'].isin(boroughs_to_keep)]
df = df.T
df
```

```python
df = df.reset_index()
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df = df.rename(columns={df.columns[0]: 'Year'})
df = df.set_index('Year')

df
```

```python
df.index = df.index.astype(int)
cleaned_df = df.loc[2017:2024]
cleaned_df
```

```python
cleaned_df.to_csv('weekly_income_cleaned.csv')
```
