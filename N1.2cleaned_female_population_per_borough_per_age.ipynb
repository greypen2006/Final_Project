```python
from matplotlib import pyplot as plt
from matplotlib import pylab
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as mcolors
import matplotlib as mpl
from matplotlib import cm
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as mpatches
```

```python
df=pd.read_csv('female pop ber borough uncleaned.csv')
df
```

```python
df = df.drop(columns=['Code'])
df = df[df['Geography'] == 'London Borough']
df
```

```python
boroughs_to_keep = ['Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich', 'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea', 'Westminster', 'Brent']
df = df[df['Name'].isin(boroughs_to_keep)]
df
```

```python
df.columns = df.columns.str.strip()
cols_to_keep = ['Name', 'All ages'] + [str(age) for age in range(18, 26)]
df = df[cols_to_keep]
df = df.rename(columns={'Name': 'Borough'})
df
```

```python
age_cols = ['All ages'] + [str(age) for age in range(18, 26)]
df[age_cols] = (df[age_cols].replace(',', '', regex=True).apply(pd.to_numeric))
df
```

```python
df.to_csv('femalepop_by_borough_and_age_cleaned.csv', index=False)
```
