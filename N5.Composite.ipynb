```python
import pandas as pd

# Open the merged dataset
file_id = "13w6neKL1JOUdgHPSyrQ7RvOD89wIJKjg"
file_url = f"https://drive.google.com/uc?id={file_id}"
df = pd.read_csv(file_url)
df.head()
```

```python
# Merge the data for each borough to get the average crime rate and rent price for each borough
df_merged = df.groupby(['BoroughName'])[['rent price', 'sexual crime']].mean().reset_index()

df_merged
```

```python
# Set the weights to 0.5 as the factors are of equal importance
wp, wc = 0.5, 0.5

# Get a normalised crime score
crime_score_calc = (
    df_merged["sexual crime"].max() - df_merged["sexual crime"]
) / (
    df_merged["sexual crime"].max() - df_merged["sexual crime"].min()
    )

df_merged["crime_score"] = crime_score_calc

df_merged
```

```python
# Get a normalised rent score
df_merged["rent_score"] = (
    df_merged["rent price"].max() - df_merged["rent price"]
) / (
    df_merged["rent price"].max() - df_merged["rent price"].min()
)

# Get a composite score
df_merged["Composite_Score"] = wp * df_merged["rent_score"] + wc * df_merged["crime_score"]

df2 = df_merged.sort_values("Composite_Score", ascending = False)

df2
```

```python
df2.to_csv('composite_score.csv')
```
